```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ===== Dados =====
data = {
    "Interventions": ["Baseline", "0", "1", "2", "3", "4", "5"],
    "RT": [6.399, 4.883, 3.789, 4.139, 3.989, 4.191, 2.883],
    "max_rt": [8.117, 6.957, 6.217, 5.957, 6.590, 6.780, 6.177],
    "MS": [2.333, 5.940, 9.410, 7.124, 8.455, 9.284, 9.238],
    "MD": [0.900, 1.788, 0.900, 0.929, 0.900, 0.900, 0.627]
}

df = pd.DataFrame(data)

# ===== Estilo =====
sns.set_theme(style="whitegrid")
plt.figure(figsize=(11, 6))

palette = {
    "RT": "#1f77b4",
    "max_rt": "#ff7f0e",
    "MS": "#2ca02c",
    "MD": "#d62728"
}

# ===== Plot =====
plt.plot(df["Interventions"], df["RT"], marker="o", linewidth=3, label="Mean RT", color=palette["RT"])
plt.plot(df["Interventions"], df["max_rt"], marker="o", linewidth=3, label="Max RT", color=palette["max_rt"])
plt.plot(df["Interventions"], df["MS"], marker="o", linewidth=3, label="Mean Active Servers (MS)", color=palette["MS"])
plt.plot(df["Interventions"], df["MD"], marker="o", linewidth=3, label="Mean Dimmer (MD)", color=palette["MD"])

# ===== Destaque visual do baseline =====
plt.axvline(x="Baseline", color="gray", linestyle="--", alpha=0.5)
plt.text(-0.05, max(df["max_rt"]) + 0.3, "Baseline", color="gray", fontsize=10, fontstyle="italic")

# ===== Labels e layout =====
plt.xlabel("Judge Interventions", fontsize=15)
plt.ylabel("Metric Value", fontsize=15)
plt.xticks(fontsize=15)
plt.yticks(fontsize=15)
plt.grid(True, alpha=0.3)

# ===== Legenda fora e acima =====
plt.legend(
    fontsize=13,
    frameon=True,
    fancybox=True,
    shadow=False,
    ncol=4,  # <-- mais horizontal
    loc="upper center",
    bbox_to_anchor=(0.5, 1.18)  # <-- fora do grid e centralizada
)

plt.tight_layout(rect=[0, 0, 1, 0.93])  # deixa espao para a legenda
plt.savefig('judges_analysis.pdf', dpi=300)
plt.show()
```
