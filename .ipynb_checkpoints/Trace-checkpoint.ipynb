```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
def read_delta_file(filename):
    data = []
    
    with open(filename, 'r') as f:
        for line in f:
            line = line.replace('\n', '')
            data.append(float(line))
    
    df = pd.DataFrame(data, columns=['value'])
    return df
```

```python
df = read_delta_file('file.delta')
df
```

```python
df['value'].sum()
```

```python
df['cumulative_time'] = df['value'].cumsum()

df['request_count'] = range(1, len(df) + 1)

df
```

```python
plt.figure(figsize=(12, 6))
plt.plot(df['cumulative_time'], df['request_count'], linewidth=2)
plt.xlabel('Tempo (segundos)')
plt.ylabel('Número de Requisições')
plt.title('Requisições Acumuladas ao Longo do Tempo')
plt.grid(True, alpha=0.3)
plt.show()
```

```python
df['cumulative_time'] = df['value'].cumsum()
df['timestamp'] = df['cumulative_time']

df['second_bucket'] = np.floor(df['timestamp']).astype(int)

requests_per_second = df['second_bucket'].value_counts().sort_index()

requests_per_second
```

```python
if len(requests_per_second) > 0:
    full_range = pd.Series(index=np.arange(requests_per_second.index.min(), 
                                         requests_per_second.index.max() + 1), 
                          data=0)
    full_range.update(requests_per_second)
    requests_per_second = full_range
```

```python
plt.figure(figsize=(14, 6))

bars = plt.bar(requests_per_second.index, requests_per_second.values, 
               alpha=0.7, color='steelblue', edgecolor='navy', linewidth=0.5)

plt.xlabel('Tempo (segundos)', fontsize=12)
plt.ylabel('Requisições por Segundo', fontsize=12)
plt.title('Throughput: Requisições por Segundo', fontsize=14, fontweight='bold')
plt.grid(True, alpha=0.3, axis='y')

mean_requests = requests_per_second.mean()
plt.axhline(y=mean_requests, color='red', linestyle='--', 
           label=f'Média: {mean_requests:.1f} req/s')
plt.legend()

if len(requests_per_second) > 30:
    plt.xticks(rotation=45)
    ticks = requests_per_second.index[::max(1, len(requests_per_second)//20)]
    plt.xticks(ticks)

plt.tight_layout()
plt.show()

print("=== ESTATÍSTICAS DE THROUGHPUT ===")
print(f"Período total: {len(requests_per_second)} segundos")
print(f"Requisições totais: {requests_per_second.sum()}")
print(f"Média: {requests_per_second.mean():.2f} req/s")
print(f"Máximo: {requests_per_second.max()} req/s")
print(f"Mínimo: {requests_per_second.min()} req/s")
print(f"Desvio padrão: {requests_per_second.std():.2f} req/s")
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(15, 10))

bars = ax1.bar(requests_per_second.index, requests_per_second.values,
               alpha=0.6, color='lightblue', edgecolor='blue', linewidth=0.5)
ax1.set_xlabel('Tempo (segundos)')
ax1.set_ylabel('Requisições por Segundo')
ax1.set_title('Throughput - Requisições por Segundo', fontsize=14, fontweight='bold')
ax1.grid(True, alpha=0.3, axis='y')
ax1.axhline(y=mean_requests, color='red', linestyle='--', 
           label=f'Média: {mean_requests:.1f} req/s')
ax1.legend()

window_size = max(1, len(requests_per_second) // 20)  # Janela adaptativa
smoothed = requests_per_second.rolling(window=window_size, center=True).mean()

ax2.plot(requests_per_second.index, requests_per_second.values, 
         alpha=0.3, color='gray', label='Original')
ax2.plot(smoothed.index, smoothed.values, 
         linewidth=2, color='darkred', label=f'Média Móvel (janela={window_size}s)')
ax2.axhline(y=mean_requests, color='red', linestyle='--', 
           label=f'Média Global: {mean_requests:.1f} req/s')
ax2.set_xlabel('Tempo (segundos)')
ax2.set_ylabel('Requisições por Segundo')
ax2.set_title('Throughput Suavizado', fontsize=14, fontweight='bold')
ax2.grid(True, alpha=0.3)
ax2.legend()

plt.tight_layout()
plt.show()
```

```python
import random
import numpy as np

def gerar_intervalos_requisicoes(num_requisicoes, tempo_total, natureza):
    natureza_normalizada = [n / 10.0 for n in natureza]
    full = []
    for p in natureza_normalizada:
        padrao_completo = ([p] * (num_requisicoes // len(natureza_normalizada) + 1))[:num_requisicoes]
        full.extend(padrao_completo)
    padrao_completo = full
    taxa_media = num_requisicoes / tempo_total
    intervalos_base = np.random.exponential(1/taxa_media, num_requisicoes)
    
    intervalos_ajustados = []
    for i, intervalo in enumerate(intervalos_base):
        fator_ajuste = 1.0 / padrao_completo[i] if padrao_completo[i] > 0 else 1.0
        intervalo_ajustado = intervalo * fator_ajuste
        intervalos_ajustados.append(intervalo_ajustado)
    
    soma_intervalos = sum(intervalos_ajustados)
    fator_normalizacao = tempo_total / soma_intervalos
    intervalos_finais = [intervalo * fator_normalizacao for intervalo in intervalos_ajustados]
    
    intervalos_finais = [max(intervalo, 0.001) for intervalo in intervalos_finais]
    
    soma_final = sum(intervalos_finais)
    if soma_final > 0:
        intervalos_finais = [intervalo * tempo_total / soma_final for intervalo in intervalos_finais]
    
    return intervalos_finais
```

```python
TEMPO = 600
REQ = 13333
```

```python
intervalos = gerar_intervalos_requisicoes(REQ, TEMPO, [1,1,2,3,4,9,10,9,9,8,10,2,3,2,1,1,3,4,1,1])
```

```python
tempos_acumulados = np.cumsum(intervalos)
requisicoes_por_segundo = [0] * int(TEMPO)

for tempo in tempos_acumulados:
    segundo = int(tempo)
    if segundo < TEMPO:
        requisicoes_por_segundo[segundo] += 1

import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))
plt.bar(range(len(requisicoes_por_segundo)), requisicoes_por_segundo, alpha=0.7)
plt.xlabel('Segundos')
plt.ylabel('Requisições por Segundo')
plt.title('Padrão de Requisições Gerado')
plt.grid(True, alpha=0.3)
plt.show()

```
